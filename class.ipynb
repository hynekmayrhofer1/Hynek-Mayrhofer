```python
class Pes:
    def __init__(self, jmeno, plemeno):
        self.jmeno = jmeno
        self.plemeno = plemeno

    def stekej(self):
        print(f"{self.jmeno} říká Haf!")

muj_pes = Pes("Buddy", "Zlatý retrívr")

print(f"Můj pes se jmenuje {muj_pes.jmeno}.")
print(f"{muj_pes.jmeno} je {muj_pes.plemeno}.")

muj_pes.stekej()

dalsi_pes = Pes("Lucy", "Labrador")
dalsi_pes.stekej()
```
