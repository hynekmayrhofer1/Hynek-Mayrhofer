```python
import random
from ColabTurtle.Turtle import *

initializeTurtle()
speed(13)
hideturtle()
pensize(3)

def random_neon():
    r = random.randint(150, 255)
    g = random.randint(80, 255)
    b = random.randint(80, 255)
    return (r, g, b)

def neon_shades(base, count):
    shades = []
    r, g, b = base

    for i in range(count):
        factor = 1 - (i * random.uniform(0.1, 0.25))  # náhodné tmavnutí
        new_r = int(max(0, min(255, r * factor)))
        new_g = int(max(0, min(255, g * factor)))
        new_b = int(max(0, min(255, b * factor)))
        shades.append((new_r, new_g, new_b))

    return shades

def chaos_triangle(x, y):

    # Náhodná velikost a počet vrstev
    size = random.randint(60, 140)
    layers = random.randint(2, 7)

    # Náhodná barva
    base_color = random_neon()
    shade_list = neon_shades(base_color, layers)

    # Náhodné pootočení
    setheading(random.randint(0, 360))

    # Náhodné posunutí
    x += random.randint(-40, 40)
    y += random.randint(-40, 40)

    # Kreslení jednotlivých vrstev
    for i, col in enumerate(shade_list):

        penup()
        goto(x, y - i * 3)
        pendown()

        pencolor(col)

        local_size = size - i * random.randint(3, 8)

        for _ in range(3):
            forward(local_size)
            left(120 + random.randint(-5, 5))  # lehce deformované úhly


center_x = 400
center_y = 250

for _ in range(100):      # počet chaotických struktur

    cx_offset = random.randint(-350, 350)

    cy_offset = random.randint(-192, 250)

    final_x = center_x + cx_offset
    final_y = center_y + cy_offset

    chaos_triangle(final_x, final_y)
```
